```python
strategies=['Rock','Paper','Scissors','Spock','Lizard']
```

```python
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0),
        ('Rock','Lizard'):(1,0),
        ('Lizard','Rock'):(0,1),
        ('Paper','Lizard'):(0,1),
        ('Lizard','Paper'):(1,0),
        ('Lizard','Scissors'):(0,1),
        ('Scissors','Lizard'):(1,0),
        ('Spock','Rock'):(1,0),
        ('Rock','Spock'):(0,1),
        ('Spock','Lizard'):(0,1),
        ('Lizard','Spock'):(1,0),
        ('Paper','Spock'):(1,0),
        ('Spock','Paper'):(0,1),
        ('Scissors','Spock'):(0,1),
        ('Spock','Scissors'):(1,0),
        ('Lizard','Lizard'):(0,0),
        ('Spock','Spock'):(0,0)}
```

```python
# resetting society
society=[{'name':n,'score':0,'strategy':None} for n in names]

# several rounds
for aRound in range(100):

    # en each round:
    for player1,player2 in itertools.combinations(society,2):
        # each chooses strategy
        player1['strategy']=choice(strategies)

        player2['strategy']=choice(strategies)

        # result from strategy chosen
        result=payoff[player1['strategy'],player2['strategy']]

        # update scores
        player1['score']+=result[0]
        player2['score']+=result[1]
```
